```python
from sim_class import Simulation
import time

# Create
sim = Simulation(num_agents=1)
print("Created")

```

```python
# Run actions
actions = [[0.1, 0.1, 0.9, 1]]
sim.run(actions, num_steps=100)
print("Actions 1 done")

time.sleep(2)

# More actions
actions = [[-0.1, -0.2, -0.9, 0]]
sim.run(actions, num_steps=100)
print("Actions 2 done")

# Reset
sim.reset(num_agents=1)
print("Reset done")

# Run again
actions = [[0.2, 0.2, 0.2, 0]]
sim.run(actions, num_steps=100)
print("Actions 3 done")

print("\nDone - restart kernel to close window")
```
